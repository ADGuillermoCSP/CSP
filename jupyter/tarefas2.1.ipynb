### Exercicio 1: Fundamentos de Clase e Obxectos

- Define unha clase chamada "Rectangulo" cun construtor que reciba os parámetros: "lonxitude" e "ancho". Gárdaos en atributos de instancia.
- Engade un método chamado "area" que calcule e devolva a área do rectángulo.
- Crea dous obxectos (instancias) de "Rectangulo" con diferentes medidas, calcula a súa área e imprime o seu diccionario interno de atributos.


```python
class Rectangulo:
    def __init__(self, lonxitude, ancho):
        # Atributos de instancia
        self.lonxitude = lonxitude
        self.ancho = ancho

    def area(self):
        # Calcula y devuelve el área
        return self.lonxitude * self.ancho


# Crear dos objetos con diferentes medidas
rectangulo1 = Rectangulo(5, 3)
rectangulo2 = Rectangulo(10, 4)

# Calcular áreas
print("Área rectángulo 1: ", rectangulo1.area())
print("Área rectángulo 2: ", rectangulo2.area())

# Imprimir diccionario interno de atributos
print("Atributos rectángulo 1: ", rectangulo1.__dict__)
print("Atributos rectángulo 2: ", rectangulo2.__dict__)
```

### Exercicio 2: Atributos "Privados" e Propiedades

- Crea unha clase "ContaBancaria".
- No constructor (`__init__`), inicializa un atributo "privado" denominado "saldo". Que non se poida acceder directamente a el.
- Crea un "getter" para poder consultar dito saldo.
- Crea un "setter" que, ao intentar modificar o saldo, só permita asignar valores positivos e "racionales".
- Fai probas de uso.

```python
class ContaBancaria:
    def __init__(self, saldo_inicial):
        # Atributo privado
        self.__saldo = 0
        self.saldo = saldo_inicial

    # Getter
    @property
    def saldo(self):
        return self.__saldo

    # Setter
    @saldo.setter
    def saldo(self, valor):
        if isinstance(valor, (int, float)) and valor >= 0:
            self.__saldo = valor
        else:
            print("Erro: o saldo debe ser un número positivo e racional.")


# Probas de uso
conta = ContaBancaria(100)

print("Saldo inicial:", conta.saldo)

# Notificación válida
conta.saldo = 250.50
print("Saldo actualizado:", conta.saldo)

# Modificación inválida (non numérico)
conta.saldo = -50

# Modificación inválida
conta.saldo = "cen"

# Intento de acceso directo
# print(conta.__saldo)  # Produce erro
```

### Exercicio 3: Herdanza Simple e Especialización

- A partires da clase "Pila" proporcionada nos contidos que ten un atributo privado "__stack_list" e métodos "push" e "pop",  crea unha nova clase denominada "PilaSegura" que herde de "Pila".
- Deberás sobrescribir o método "pop" para que, antes de devolvelo, imprima unha mensaxe coma "Elemento retirado correctamente".
- Engade un novo método "esta_baleira" que devuelva Verdadeiro ou Falso dependendo de se a pila ten ou non elementos.
- Utiliza métodos coma "issubclass" e "isinstance" para comprobar a relación.


```python
class Pila:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        if len(self.__stack_list) > 0:
            return self.__stack_list.pop()
        else:
            return None


class PilaSegura(Pila):
    def pop(self):
        elemento = super().pop()
        if elemento is not None:
            print("Elemento retirado correctamente")
        return elemento

    def esta_baleira(self):
        elemento = super().pop()
        if elemento is None:
            return True
        else:
            super().push(elemento)
            return False


pila = PilaSegura()

print("Está baleira?", pila.esta_baleira())

pila.push(10)
pila.push(20)

print("Está baleira?", pila.esta_baleira())

print("Elemento extraído:", pila.pop())

print("Está baleira?", pila.esta_baleira())

print("PilaSegura é subclase de Pila?", issubclass(PilaSegura, Pila))
print("pila é instancia de PilaSegura?", isinstance(pila, PilaSegura))
print("pila é instancia de Pila?", isinstance(pila, Pila))
```

### Exercicio 4: Atributos de Clase vs. Instancia

- Crea unha clase "Coche" cun atributo de clase denominado "numero_de_rodas" e con valor = 4.
- No constructor define un atributo de instancia denominado "marca" que se reciba coma parámetro.
- Ten tamén un atributo de clase denominado "total_coches_creados" que leve a conta de todas as instancias xeradas.
- Crea 5 obxectos "Coche" de distintas marcas.
- Imprime, en cada caso, os atributos de clase e o diccionario de cada instancia.


```python
class Coche:
    # Atritubos de clase
    numero_de_rodas = 4
    total_coches_creados = 0

    def __init__(self, marca):
        # Atributo de instancia
        self.marca = marca

        # Incrementar contador de coches creados
        Coche.total_coches_creados += 1


# Crear 5 coches
coche1 = Coche("Toyota")
coche2 = Coche("Ford")
coche3 = Coche("BMW")
coche4 = Coche("Audio")
coche5 = Coche("Tesla")

coches = [coche1, coche2, coche3, coche4, coche5]

# Mostrar información
for coche in coches:
    print("Marca:", coche.marca)
    print("Número de rodas:", Coche.numero_de_rodas)
    print("Total de coches creados:", Coche.total_coches_creados)
    print("Diccionario da instancia", coche.__dict__)
```

### Exercicio 5: Métodos Estáticos, de Clase e de Instancia

- Crea unha clase denominada "UtilidadesMatematicas" cos seguintes métodos:
    - Un método estático denominado "e_par" que devolva True ou False.
    - Un método de clase denominado "informacion" que devolva unha cadea descritiva sobre a clase.
    - Un método de instancia denominado "potencia" que, para un número base almacenado nun atributo de instancia inicializado no constructor, calcule a potencia a dito exponente.
- Fai o preciso para chamar a todos os métodos creados.

```python
class UtilidadesMatematicas:
    def __init__(self, base):
        self.base = base

    @staticmethod
    def e_par(numero):
        return numero % 2 == 0

    @classmethod
    def informacion(cls):
        return f"A clase {cls.__name__} contén utilidades matemáticas"

    def potencia(self, exponente):
        return self.base ** exponente


print("¿10 é par?", UtilidadesMatematicas.e_par(10))
print("¿7 é par?", UtilidadesMatematicas.e_par(7))

print(UtilidadesMatematicas.informacion())

util = UtilidadesMatematicas(3)
print("Potencia:", util.potencia(4))
```

### Exercicio 6: Dunder Methods

- Crea unha clase "Vector2D" para representar un vector no plano cun constructor que reciba as coordenadas "x" e "y".
- Implementa (ou sobrescribe) o método "__add__" para que poidas sumar dous vectores usando o operador "+". Deberá devolver un obxecto novo coa suma das coordenadas.
- Implementa (ou sobrescribe) o método "__str__" para que, cando se convirta o vector a string, amose un formato lexible como "Vector2D(x=5, y=3)".
- Implementa (ou sobrescribe) o método "__eq__" para que dous vectores sexan iguais se as súas coordenadas "x" e "y" coinciden.
- Constrúe exemplos de uso.

```python
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Suma de vectores con +
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)

    # Representación en texto
    def __str__(self):
        return f"Vector2D(x={self.x}, y={self.y})"

    # Comparación de igualdad
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


v1 = Vector2D(5, 3)
v2 = Vector2D(2, 4)

# Suma de vectores
v3 = v1 + v2

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Suma:", v3)

# Comparación
v4 = Vector2D(7, 7)

print("v3 == v4 ?", v3 == v4)
print("v1 == v2 ?", v1 == v2)
```

### Exercicio 7: Herdanza Múltiple e Dunder Methods

- Define unha clase simple "Imprimible" co método "amosar" que devolva "Son un obxecto imprimible".
- A partires da clase anterior "Pila", haberá outra que se chame "PilaSuma" que herde de "Pila". crea unha nova clase "PilaSumaImprimible" que herde de ambas: "PilaSuma" e "Imprimible".
- No constructor de "PilaSumaImprimible" inicializa correctamente a parte correspondente de "PilaSuma".
- Sobrescribe o método "amosar" para que, no lugar do texto xenérico, devolva un string informativo como "PilaSuma coa suma total: X", onde "X" será o valor da propiedad "suma".
- Crea unha instancia desta clase, engade algúns números co "push", e proba a chamar tanto aos métodos herdados ("pop" e "suma") como ao método personalizado "amosar".
- Que fai "__mro__"? E "__bases__"? Utiliza métodos como issubclass.

```python
class Pila:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        if len(self.__stack_list) > 0:
            return self.__stack_list.pop()
        else:
            return None


class PilaSuma(Pila):
    def __init__(self):
        super().__init__()
        self.suma = 0

    def push(self, val):
        super().push(val)
        self.suma += val

    def pop(self):
        val = super().pop()
        if val is not None:
            self.suma -= val
        return val


class Imprimible:
    def amosar(self):
        return "Son un obxecto imprimible"


class PilaSumaImprimible(PilaSuma, Imprimible):
    def __init__(self):
        super().__init__()

    def amosar(self):
        return f"PilaSuma con suma total: {self.suma}"


pila = PilaSumaImprimible()

pila.push(10)
pila.push(5)
pila.push(7)

print("Suma actual:", pila.suma)

print("Pop:", pila.pop())

print(pila.amosar())

# MRO significa Methor Resolution Order y es el orden en el que Python busca métodos
print("MRO:", PilaSumaImprimible.__mro__)
# __bases__ devuelde las clases padres directas
print("Bases:", PilaSumaImprimible.__bases__)

print("É subclase de Pila?", issubclass(PilaSumaImprimible, Pila))
print("É subclase de Imprimible?", issubclass(PilaSumaImprimible, Imprimible))
print("É instancia?", isinstance(Pila, PilaSumaImprimible))
```
