### Apertura Segura con try...except:

* Escribe unha función chamada `abrir_ficheiro_seguro(nome_ficheiro, modo)` que intente abrir un ficheiro no modo especificado.
* Debe usar un bloque `try...except` para capturar calquera excepción (como `FileNotFoundError`).
* Se a apertura é exitosa, a función debe devolver o obxecto do ficheiro. Se hai un erro, debe amosar a mensaxe de erro e devolver `None`.
* Fai probas invocando a función cun ficheiro que non exista e despois cun que si exista (por exemplo, en modo lectura `"r"`). Non esquezas pechar o ficheiro se se abriu correctamente.

### Diario de Notas Persoal

* Crea un programa que permita ao usuario escribir anotacións nun ficheiro chamado `diario.txt`.
* O programa debe:
    1. Preguntar ao usuario se quere (1) Ler todas as anotacións ou (2) Engadir unha nova anotación.
    2. Se escolle ler, debe abrir o ficheiro en modo lectura (`"r"`) e amosar todo o seu contido na pantalla. Controla o erro se o ficheiro aínda non existe.
    3. Se escolle engadir, debe pedirlle ao usuario unha liña de texto e **engadila** ao final do ficheiro (modo `"a"`), incluíndo a data e hora actual ao comezo da liña (usa o módulo `datetime`).


````python
import datetime

class DiarioPersoal:
    def __init__(self, nome_ficheiro="diario.txt"):
        self.__nome_ficheiro = nome_ficheiro

    def ler_anotacions(self):
        try:
            with open(self.__nome_ficheiro, "r", encoding="utf-8") as ficheiro:
                contido = ficheiro.read()
                if not contido:
                    print("\n--- O diario está baleiro polo de agora. ---")
                else:
                    print("\n--- CONTIDO DO DIARIO ---")
                    print(contido)
        except FileNotFoundError:
            print(f"\n[ERRO]: O ficheiro '{self.__nome_ficheiro}' aínda non existe. Engade unha nota primeiro.")
        except Exception as e:
            print(f"\n[ERRO INESPERADO]: {e}")

    def engadir_anotacion(self, texto):
        agora = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        linea_completa = f"[{agora}] - {texto}\n"

        try:
            with open(self.__nome_ficheiro, "a", encoding="utf-8") as ficheiro:
                ficheiro.write(linea_completa)
            print("Anotación gardada correctamente.")
        except Exception as e:
            print(f"\n[ERRO AO GARDAR]: {e}")

def executar_programa():
    diario = DiarioPersoal()

    print("BENVIDO AO TEU DIARIO PERSOAL")
    print("1. Ler todas as anotacións")
    print("2. Engadir unha nova anotación")

    opcion = input("\nEscolle unha opción (1 ou 2): ").strip()

    if opcion == "1":
        diario.ler_anotacions()
    elif opcion == "2":
        nova_nota = input("Escribe a túa anotación: ")
        diario.engadir_anotacion(nova_nota)
    else:
        print("Opción non válida. Por favor, executa a celda de novo e elixe 1 ou 2.")

# Execución do programa
if __name__ == "__main__":
    executar_programa()
````

### Analizador de Ficheiro de Texto

* Pídelle ao usuario o nome dun ficheiro de texto.
* O programa debe abrir o ficheiro e calcular:
    * Número total de liñas.
    * Número total de palabras.
    * Número total de caracteres.
* Crea un ficheiro `proba.txt` con algunhas liñas de texto e comproba que os cálculos son correctos.

````python
def analizar_ficheiro(nome_ficheiro):
    try:
        with open(nome_ficheiro, 'r', encoding='utf-8') as arquivo_lectura:
            lineas = arquivo_lectura.readlines()

            num_lineas = len(lineas)
            # Calculamos palabras e caracteres percorrendo as liñas
            num_palabras = sum(len(linea.split()) for linea in lineas)
            num_caracteres = sum(len(linea) for linea in lineas)

            return num_lineas, num_palabras, num_caracteres

    except FileNotFoundError:
        print(f"Erro: O ficheiro '{nome_ficheiro}' non existe.")
        return None
    except Exception as e:
        print(f"Produciuse un erro inesperado: {e}")
        return None

# Creación do ficheiro 'proba.txt' para o test
contido_proba = """Ola, isto é unha proba.
Segunda liña do ficheiro.
Fin do analizador."""

with open("proba.txt", "w", encoding="utf-8") as f_proba:
    f_proba.write(contido_proba)

# Interacción co usuario e execución
nome_usuario = input("Introduce o nome do ficheiro a analizar (ex: proba.txt): ").strip()

resultados = analizar_ficheiro(nome_usuario)

if resultados:
    l, p, c = resultados
    print(f"Resultados para '{nome_usuario}':")
    print(f"  * Número total de liñas: {l}")
    print(f"  * Número total de palabras: {p}")
    print(f"  * Número total de caracteres: {c}")
````

### Copiador de Ficheiros con Número de Liña

* Pídelle ao usuario o nome dun ficheiro de orixe.
* Crea un novo ficheiro denominado `copia_con_numeros.txt`.
* Le o ficheiro de orixe **liña por liña** e escribe cada liña no novo ficheiro, pero **engadíndolle o número de liña ao principio**.
* Por exemplo, se unha liña orixinal é "Hola mundo", no novo ficheiro debe aparecer "1 Hola mundo".

````python
import os

def copiar_con_numeracion(ficheiro_orixe):
    ficheiro_destino = "copia_con_numeros.txt"

    try:
        with open(ficheiro_orixe, 'r', encoding='utf-8') as orixe, \
             open(ficheiro_destino, 'w', encoding='utf-8') as destino:

            # Iteramos liña a liña
            for i, linea in enumerate(orixe, start=1):
                # Escribimos o número seguido do contido da liña
                destino.write(f"{i} {linea}")

        print(f"\nÉxito: Creouse '{ficheiro_destino}' con números de liña.")

    except FileNotFoundError:
        print(f"\n[ERRO]: O ficheiro '{ficheiro_orixe}' non existe.")
    except Exception as e:
        print(f"\n[ERRO INESPERADO]: {e}")


# Creamos un ficheiro de proba
proba_nome = "orixe_test.txt"
with open(proba_nome, "w", encoding="utf-8") as f:
    f.write("Primeira liña de exemplo.\nSegunda liña de texto.\nÚltima liña do ficheiro.")

print(f"COPIADOR NUMERADO")

# Pedimos o nome do ficheiro ao usuario
nome_input = input(f"\nIntroduce o nome do ficheiro de orixe (ex: {proba_nome}): ").strip()

# Executamos a función
copiar_con_numeracion(nome_input)

# Mostrar resultado se todo foi ben
if os.path.exists("copia_con_numeros.txt") and nome_input == proba_nome:
    print("\nContido da copia:")
    with open("copia_con_numeros.txt", "r") as f:
        print(f.read())
````

### Estracto bancario en CSV

* Crea un programa para ler e almacenar un estracto bancario do teu banco en CSV.
* Os pasos que terás que dar son os seguintes:
    * Descarga un ficheiro CSV do teu banco.
    * Instala `chardet` para detectar o encoding do teu ficheiro.
    * Averigua as cabeceiras. Por exemplo: `Data ctble;Data valor;Concepto;Importe;Moeda;Saldo;Moeda;Concepto ampliado`. Un exemplo sería: `02-01-2026;29-12-2025;123456789123 PEIXERIA TRINQUETE 123456789;-14,97;EUR;1234,62;EUR;`
    * O programa debe ter un menú para:
        * Cargar os datos dun CSV nun Diccionario ou, mellor dito, nunha lista de Diccionarios.
        * Buscar por concepto e amosar liña a liña todos os conceptos que conteñan a cadea introducida.
        * Buscar por cantidade e amosar liña a liña todas as cantidades que coincidan.
        * Suma por conceptos. Se lle introduce un concepto e amosa a suma de todas as entradas que conteñan ese concepto.
        * Procurar movementos entre datas e amosalos por pantalla.
    * O programa só pode rematar cando se lle indique a través dunha opción de menú.

````python
import chardet
import csv
from datetime import datetime

class XestorEstracto:
    def __init__(self):
        self.datos = []
        self.cabeceiras = []

    @staticmethod
    def detectar_encoding(ruta):
        with open(ruta, 'rb') as archivo_bin:
            resultado = chardet.detect(archivo_bin.read(10000))
        return resultado['encoding']

    def cargar_csv(self, ruta):
        try:
            encoding = self.detectar_encoding(ruta)
            with open(ruta, mode='r', encoding=encoding) as f_csv:
                lector = csv.DictReader(f_csv, delimiter=';')
                self.datos = list(lector)
                self.cabeceiras = lector.fieldnames
            print(f"\n[OK] Cargados {len(self.datos)} movementos.")
        except FileNotFoundError:
            print("\n[ERRO] Non se atopou o ficheiro CSV.")
        except Exception as e:
            print(f"\n[ERRO] Problema ao cargar: {e}")

    def buscar_por_concepto(self, cadea):
        print(f"\n--- Resultados para concepto: '{cadea}' ---")
        encontrados = [d for d in self.datos if cadea.lower() in d['Concepto'].lower()]
        for r in encontrados:
            print(r)
        return encontrados

    def buscar_por_cantidade(self, cantidade):
        print(f"\n--- Resultados para importe: {cantidade} ---")
        encontrados = [d for d in self.datos if d['Importe'].replace(',', '.') == str(cantidade).replace(',', '.')]
        for r in encontrados:
            print(r)

    def sumar_por_concepto(self, cadea):
        encontrados = [d for d in self.datos if cadea.lower() in d['Concepto'].lower()]
        total = sum(float(d['Importe'].replace(',', '.')) for d in encontrados)
        print(f"\nSuma total para '{cadea}': {total:.2f} EUR")

    def filtrar_por_datas(self, inicio, fin):
        formato = "%d-%m-%Y"
        d_inicio = datetime.strptime(inicio, formato)
        d_fin = datetime.strptime(fin, formato)

        print(f"\n--- Movementos entre {inicio} e {fin} ---")
        for d in self.datos:
            data_mov = datetime.strptime(d['Data ctble'], formato)
            if d_inicio <= data_mov <= d_fin:
                print(d)

def menu():
    xestor = XestorEstracto()

    # Creamos un ficheiro de exemplo
    with open("estracto_proba.csv", "w", encoding="utf-8") as ficheiro_saida:
        ficheiro_saida.write("Data ctble;Data valor;Concepto;Importe;Moeda;Saldo;Moeda;Concepto ampliado\n")
        ficheiro_saida.write("02-01-2026;29-12-2025;PEIXERIA TRINQUETE;-14,97;EUR;1234,62;EUR;\n")
        ficheiro_saida.write("05-01-2026;05-01-2026;NOMINA EMPRESA;1500,00;EUR;2734,62;EUR;\n")
        ficheiro_saida.write("10-01-2026;10-01-2026;SUPERMERCADO;-45,20;EUR;2689,42;EUR;\n")

    while True:
        print("\n=== MENÚ ESTRACTO BANCARIO ===")
        print("1. Cargar datos CSV")
        print("2. Buscar por concepto")
        print("3. Buscar por cantidade")
        print("4. Suma por concepto")
        print("5. Buscar entre datas (dd-mm-aaaa)")
        print("6. Saír")

        opcion = input("\nSelecciona unha opción: ")

        if opcion == "1":
            ruta = input("Nome do ficheiro (ex: estracto_proba.csv): ")
            xestor.cargar_csv(ruta)
        elif opcion == "2":
            conc = input("Introduce a cadea a buscar: ")
            xestor.buscar_por_concepto(conc)
        elif opcion == "3":
            cant = input("Introduce o importe exacto: ")
            xestor.buscar_por_cantidade(cant)
        elif opcion == "4":
            conc = input("Concepto para sumar: ")
            xestor.sumar_por_concepto(conc)
        elif opcion == "5":
            ini = input("Data inicio (dd-mm-aaaa): ")
            fin = input("Data fin (dd-mm-aaaa): ")
            try:
                xestor.filtrar_por_datas(ini, fin)
            except ValueError:
                print("[ERRO] Formato de data incorrecto.")
        elif opcion == "6":
            print("Saíndo do programa...")
            break
        else:
            print("Opción non válida.")

if __name__ == "__main__":
    menu()
````

````python
import chardet
import pandas as pd

class XestorEstracto:

    def __init__(self):
        self.df = None

    @staticmethod
    def detectar_encoding(ruta):
        with open(ruta, 'rb') as archivo_bin:
            resultado = chardet.detect(archivo_bin.read(10000))
        return resultado['encoding']

    def cargar_csv(self, ruta):
        try:
            encoding = self.detectar_encoding(ruta)

            self.df = pd.read_csv(
                ruta,
                delimiter=';',
                encoding=encoding
            )

            print(f"\n[OK] Cargados {len(self.df)} movementos.")

        except FileNotFoundError:
            print("\n[ERRO] Non se atopou o ficheiro CSV.")
        except Exception as e:
            print(f"\n[ERRO] Problema ao cargar: {e}")

    def buscar_por_concepto(self, cadea):

        if self.df is None:
            print("[ERRO] Primeiro debes cargar o CSV.")
            return

        movements = self.df[self.df["Concepto"].str.contains(cadea, case=False, na=False)]

        print(f"\n--- Resultados para concepto '{cadea}' ---")
        print(movements)

    def buscar_por_cantidade(self, cantidade):

        if self.df is None:
            print("[ERRO] Primeiro debes cargar o CSV.")
            return

        importe = str(cantidade).replace(',', '.')

        df_tmp = self.df.copy()
        df_tmp["Importe"] = df_tmp["Importe"].str.replace(',', '.').astype(float)

        resultados_importe = df_tmp[df_tmp["Importe"] == float(importe)]

        print(f"\n--- Resultados para importe {cantidade} ---")
        print(resultados_importe)

    def sumar_por_concepto(self, cadea):

        if self.df is None:
            print("[ERRO] Primeiro debes cargar o CSV.")
            return

        df_tmp = self.df.copy()
        df_tmp["Importe"] = df_tmp["Importe"].str.replace(',', '.').astype(float)

        resultados_concepto = df_tmp[df_tmp["Concepto"].str.contains(cadea, case=False, na=False)]

        total = resultados_concepto["Importe"].sum()

        print(f"\nSuma total para '{cadea}': {total:.2f} EUR")

    def filtrar_por_datas(self, inicio, fin):

        if self.df is None:
            print("[ERRO] Primeiro debes cargar o CSV.")
            return

        df_tmp = self.df.copy()

        df_tmp["Data ctble"] = pd.to_datetime(df_tmp["Data ctble"], format="%d-%m-%Y")

        inicio = pd.to_datetime(inicio, format="%d-%m-%Y")
        fin = pd.to_datetime(fin, format="%d-%m-%Y")

        resultados_fechas = df_tmp[
            (df_tmp["Data ctble"] >= inicio) &
            (df_tmp["Data ctble"] <= fin)
        ]

        print(f"\n--- Movementos entre {inicio.date()} e {fin.date()} ---")
        print(resultados_fechas)


def menu():

    xestor = XestorEstracto()

    # Creamos un CSV de exemplo
    with open("estracto_proba.csv", "w", encoding="utf-8") as ficheiro_saida:
        ficheiro_saida.write("Data ctble;Data valor;Concepto;Importe;Moeda;Saldo;Moeda;Concepto ampliado\n")
        ficheiro_saida.write("02-01-2026;29-12-2025;PEIXERIA TRINQUETE;-14,97;EUR;1234,62;EUR;\n")
        ficheiro_saida.write("05-01-2026;05-01-2026;NOMINA EMPRESA;1500,00;EUR;2734,62;EUR;\n")
        ficheiro_saida.write("10-01-2026;10-01-2026;SUPERMERCADO;-45,20;EUR;2689,42;EUR;\n")

    while True:

        print("\n=== MENÚ ESTRACTO BANCARIO ===")
        print("1. Cargar datos CSV")
        print("2. Buscar por concepto")
        print("3. Buscar por cantidade")
        print("4. Suma por concepto")
        print("5. Buscar entre datas (dd-mm-aaaa)")
        print("6. Saír")

        opcion = input("\nSelecciona unha opción: ")

        if opcion == "1":
            ruta = input("Nome do ficheiro (ex: estracto_proba.csv): ")
            xestor.cargar_csv(ruta)

        elif opcion == "2":
            conc = input("Introduce a cadea a buscar: ")
            xestor.buscar_por_concepto(conc)

        elif opcion == "3":
            cant = input("Introduce o importe exacto: ")
            xestor.buscar_por_cantidade(cant)

        elif opcion == "4":
            conc = input("Concepto para sumar: ")
            xestor.sumar_por_concepto(conc)

        elif opcion == "5":
            ini = input("Data inicio (dd-mm-aaaa): ")
            fin = input("Data fin (dd-mm-aaaa): ")

            try:
                xestor.filtrar_por_datas(ini, fin)
            except ValueError:
                print("[ERRO] Formato de data incorrecto.")

        elif opcion == "6":
            print("Saíndo do programa...")
            break

        else:
            print("Opción non válida.")


if __name__ == "__main__":
    menu()
````

###  Xestor de Contactos en JSON

* Crea un programa para xestionar unha axenda de contactos nun ficheiro de texto `contactos.JSON`.
* A estrutura será unha lista de dicionarios, onde cada dicionario ten as chaves `nome`, `telefono`, `email`.
* O programa debe ter un menú para:
    * **Engadir contacto**: Pide os datos e engádeos ao ficheiro.
    * **Buscar contacto**: Pide un nome e amosa o teléfono e email de todas as liñas que conteñan ese nome.
    * **Listar todos**: Amosa todos os contactos numerados.

````python
import json
import os

class XestorContactos:
    def __init__(self, ficheiro="contactos.json"):
        self.ficheiro = ficheiro
        if not os.path.exists(self.ficheiro):
            self._gardar([])

    def _cargar(self):
        try:
            with open(self.ficheiro, 'r', encoding='utf-8') as f:
                return json.load(f)
        except (json.JSONDecodeError, FileNotFoundError):
            return []

    def _gardar(self, contactos):
        with open(self.ficheiro, 'w', encoding='utf-8') as f:
            json.dump(contactos, f, indent=4, ensure_ascii=False)

    def engadir(self, nome, telefono, email):
        contactos = self._cargar()
        novo = {"nome": nome, "telefono": telefono, "email": email}
        contactos.append(novo)
        self._gardar(contactos)
        print(f"\n[OK] Contacto '{nome}' engadido con éxito.")

    def buscar(self, termo):
        contactos = self._cargar()
        print(f"\n--- Resultados para '{termo}' ---")
        encontrados = [c for c in contactos if termo.lower() in c['nome'].lower()]

        if not encontrados:
            print("Non se atopou ningún contacto con ese nome.")
        else:
            for c in encontrados:
                print(f"Nome: {c['nome']} | Tel: {c['telefono']} | Email: {c['email']}")

    def listar(self):
        contactos = self._cargar()
        print("\n--- LISTA DE CONTACTOS ---")
        if not contactos:
            print("A axenda está baleira.")
        else:
            for i, c in enumerate(contactos, start=1):
                print(f"{i}. {c['nome']} (Telf: {c['telefono']}, Email: {c['email']})")

def menu():
    xestor = XestorContactos()

    while True:
        print("\n--- XESTOR DE CONTACTOS JSON ---")
        print("1. Engadir contacto")
        print("2. Buscar contacto")
        print("3. Listar todos")
        print("4. Saír")

        opcion = input("\nEscolle unha opción: ").strip()

        if opcion == "1":
            n = input("Nome: ")
            t = input("Teléfono: ")
            e = input("Email: ")
            xestor.engadir(n, t, e)
        elif opcion == "2":
            termo = input("Nome a buscar: ")
            xestor.buscar(termo)
        elif opcion == "3":
            xestor.listar()
        elif opcion == "4":
            print("Pechando axenda...")
            break
        else:
            print("Opción non válida.")

if __name__ == "__main__":
    menu()
````

### Cálculo con Datos Numéricos usando NumPy

* Crea un ficheiro de texto chamado `matriz.txt` con varias liñas de números separados por espazos. Por exemplo:
    ```
    1.5 2.3 4.8
    5.1 3.2 7.4
    9.6 8.7 6.5
    ```
* Escribe un script que:
    1. Cargue os datos do ficheiro nun array de **NumPy**.
    2. Calcule e amose por pantalla: a media de todos os números, a suma de cada fila e o valor máximo de cada columna. 
    3. Finalmente, garde o array orixinal multiplicado por 2 nun novo ficheiro `matriz_dobre.txt` con formato `'%.2f'`.

````python
import numpy as np
import os


contido = """1.5 2.3 4.8
5.1 3.2 7.4
9.6 8.7 6.5"""

with open("matriz.txt", "w", encoding="utf-8") as f:
    f.write(contido)

print("Ficheiro 'matriz.txt' creado.")

try:
    # Cargamos os datos nun array de NumPy
    matriz = np.loadtxt("matriz.txt")
    print("\nMatriz Cargada")
    print(matriz)

    # Media de todos os números
    media_total = np.mean(matriz)

    # Suma de cada fila
    suma_filas = np.sum(matriz, axis=1)

    # Valor máximo de cada columna
    max_columnas = np.max(matriz, axis=0)

    # Amosar resultados
    print(f"\nMedia de todos os números: {media_total:.2f}")
    print(f"Suma de cada fila: {suma_filas}")
    print(f"Valor máximo de cada columna: {max_columnas}")

    # Gardar o array multiplicado por 2
    matriz_dobre = matriz * 2
    np.savetxt("matriz_dobre.txt", matriz_dobre, fmt='%.2f')

    print("\nFicheiro 'matriz_dobre.txt' gardado con éxito.")

except Exception as e:
    print(f"\nProduciuse un problema: {e}")

# Comprobación do ficheiro xerado
if os.path.exists("matriz_dobre.txt"):
    print("\nContido de 'matriz_dobre.txt':")
    with open("matriz_dobre.txt", "r") as f_dobre:
        print(f_dobre.read())
````
