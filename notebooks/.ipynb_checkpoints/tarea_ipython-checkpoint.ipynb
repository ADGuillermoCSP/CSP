```python
# Comprobamos versiones de Python y bibliotecas esenciales
import sys
import platform
print("Python:", sys.version.splitlines()[0])
print("Plataforma:", platform.platform())

import importlib
for pkg in ("ipython", "numpy", "matplotlib", "pandas"):
    try:
        m = importlib.import_module(pkg)
        print(f"{pkg}:", getattr(m, "__version__", "versión desconocida"))
    except Exception as e:
        print(f"{pkg}: NO INSTALADO ({e})")
```

```python
# %time: tiempo de ejecución de la última expresión
def f(n):
    s = 0
    for i in range(n):
        s += i
    return s

%time f(100000)

# %timeit: medida repetida
%timeit f(100000)
```

```python
import numpy as np

arr = np.linspace(0, 2*np.pi, 200)
print("media:", arr.mean())
print("desviación:", arr.std())
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 300)
y = np.sin(x) * np.exp(-x*0.1)

plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.title("Seno amortiguado")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
