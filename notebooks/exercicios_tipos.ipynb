### Exercicio 6: Diccionarios

A partires do seguinte diccionario:

```python
estudante = {
    "nome": "Carlos",
    "idade": 35,
    "modulos": ["Redes", "Sistemas", "Programación"],
    "activo": True
}
```

Realiza estas operacións:

1. Engade a chave "promedio" con valor 8.5
2. Actualiza la idade a 40
3. Engade "Hardware" á lista de módulos
4. Amosa todas as chaves do diccionario
5. Crea unha copia do diccionario e modifica a copia

```python
from copy import deepcopy

# 1. Engadir chave "promedio"
estudante["promedio"] = 8.5

# 2. Actualizar idade
estudante["idade"] = 40

# 3. Engadir "Hardware" á lista de módulos
estudante["modulos"].append("Hardware")

# 4. Amosar todas as chaves
print("Claves:", estudante.keys())

# 5. Crear unha copia profunda e modificarla
estudante_copia = deepcopy(estudante)
estudante_copia["modulos"].append("Seguridade")
estudante_copia["idade"] = 50

print("Orixinal:", estudante)
print("Copia modificada:", estudante_copia)
```

### Exercicio 7: Lista de Diccionarios

Crea unha lista de diccionarios con información de estudantes:

```python
estudantes = [
    {"nome": "Ana", "idade": 20, "nota": 8.5},
    {"nome": "Carlos", "idade": 22, "nota": 7.8},
    {"nome": "María", "idade": 21, "nota": 9.2}
]
```

Realiza estas operacións:

1. Engade un nopvo estudante.
2. Atopa o estudiante con maior nota.
3. Calcula o promedio de idades.
4. Crea unha lista só cos nomes.
5. Ordena os estudantes por nota (descendente).

```python
estudantes.append({"nome": "Laura", "idade": 23, "nota": 9.0})

mellor = max(estudantes, key=lambda e: e["nota"])
print("Mellor estudante:", mellor)

prom_idades = sum(e["idade"] for e in estudantes) / len(estudantes)
print("Promedio de idades:", prom_idades)

so_nomes = [e["nome"] for e in estudantes]
print("Nomes:", so_nomes)

ordenados = sorted(estudantes, key=lambda e: e["nota"], reverse=True)
print("Ordenados por nota:", ordenados)
```

### Exercicio 8: Conxuntos

A partir destes conxuntos:

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}
```

Realiza estas operacións:

1. Unión de ambos conxuntos
2. Intersección
3. Diferencia (A - B)
4. Diferencia simétrica
5. Verifica si 3 está en A

```python
print("Unión:", conx_a | conx_b)
print("Intersección:", conx_a & conx_b)
print("Diferencia:", conx_a - conx_b)
print("Diferencia simétrica:", conx_a ^ conx_b)
print("3 está en conx_a:", 3 in conx_a)
```

### Exercicio 9: Programa Integrador

Crea un programa que:

1. Pida ao usuario o seu nome, idade e cidade de residencia
2. Almacene esa información nun diccionario
3. Convirta a idade a enteiro
4. Cree unha lista de hobbies (pregunta por 3 polo menos)
5. Amose toda a información formateada de xeito lexible

```python
nome = input("Nome: ")
idade = int(input("Idade: "))
cidade = input("Cidade de residencia: ")

hobbies = []
for i in range(3):
    h = input(f"Hobby {i+1}: ")
    hobbies.append(h)

info = {
    "nome": nome,
    "idade": idade,
    "cidade": cidade,
    "hobbies": hobbies
}

print("\n--- INFORMACIÓN FINAL ---")
for k, v in info.items():
    print(f"{k}: {v}")
```

### Exercicio 10: Combinación de Tipos Complexos

Temos un Sistema de biblioteca:

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}
```

Realiza estas operacións:

1. Engade un novo libro.
2. Marca un libro como non dispoñible.
3. Atopa todos os libros dun xénero.
4. Libros dispoñibles por autor.
5. Usuario con máis libros prestados.

```python
# 1. Engadir libro
novo = {
    "titulo": "El Hobbit",
    "autor": "J.R.R. Tolkien",
    "ano": 1937,
    "xeneros": {"Fantasia"},
    "disponible": True
}
biblioteca["libros"].append(novo)

# 2. Marcar un libro como non dispoñible
for libro in biblioteca["libros"]:
    if libro["titulo"] == "Cien años de soledad":
        libro["disponible"] = False

# 3. Atopa libros dun xénero
xenero_buscado = "Distopía"
por_xenero = [l for l in biblioteca["libros"] if xenero_buscado in l["xeneros"]]
print("Libros do xénero", xenero_buscado, ":", por_xenero)

# 4. Libros dispoñibles por autor
autor = "George Orwell"
dispo_autor = [l for l in biblioteca["libros"] if l["autor"] == autor and l["disponible"]]
print("Libros dispoñibles de", autor, ":", dispo_autor)

# 5. Usuario con máis libros prestados
usuario_max = max(biblioteca["usuarios"].values(), key=lambda u: len(u["libros_prestados"]))
print("Usuario con máis libros prestados:", usuario_max)
```
