### Exercicio 1: Operadores Aritméticos

Calcula a área e perímetro dun rectángulo:
- Pide ao usuario base e altura.
- Calcula área.
- Calcula perímetro.
- Utiliza `//` e `%`.
- Amosa os resultados utilizando diferentes formatos de cadeas.

```python
base = int(input("Introduce la base del rectángulo: "))
altura = int(input("Introduce la altura del rectángulo: "))

area = base * altura
perimetro = 2 * (base + altura)

print(f"Área: {area}, perímetro: {perimetro}")
print("Área (división entera entre 2):", area//2)
print("Área módulo 3:", area%3)
```

### Exercicio 2: Operadores de Asignación

Simula unha conta bancaria:
- Saldo inicial: 1000
- Usa += para depositar 500
- Usa -= para retirar 300
- Usa *= para aplicar interés do 10%
- Usa /= para dividir entre 2 (metade para aforros)
- Amosa saldo despois de cada operación
- Utiliza diferentes formatos de cadeas.

```python
saldo = 1000
print(f"Saldo inicial: {saldo}")

saldo += 500
print(f"Después de depositar 500: {saldo}")

saldo -= 300
print(f"Después de retirar 300: {saldo}")

saldo *= 1.10
print(f"Después de aplicar 10% de interés: {saldo:.2f}")

saldo /= 2
print(f"Mitad para aforros: {saldo:.2f}")
```

### Exercicio 4: Operadores Lóxicos

Determina se un ano é bisesto:
- Divisible entre 4.
- Non divisible entre 100, excepto se é divisible entre 400.
- Non encadenes condicionais senón condicións.
- Exemplo: 2024 (bisesto), 1900 (non bisesto), 2000 (bisesto).

```python
ano = int(input("Introduce un ano: "))

if (ano % 4 == 0) and (ano % 100 != 0 or ano % 400 == 0):
    print(f"{ano} é bisesto")
else:
    print(f"{ano} non é bisesto")
```

### Exercicio 5: Operadores de Identidade

Analiza diferenzas entre == e is:
- Crea dúas listas (lista1, lista2) iguais: [1, 2, 3]
- Crea unha variable (lista3) que apunte lista2, por exemplo.
- Explica resultados de comparar `==` e `is`.

```python
lista1 = [1,2,3]
lista2 = [1,2,3]
lista3 = lista2

print("lista1 == lista2:", lista1 == lista2)  # True, valores iguais
print("lista1 is lista2:", lista1 is lista2)  # False, diferentes obxectos
print("lista2 is lista3:", lista2 is lista3)  # True, mesma referencia
```

### Exercicio 6: Operadores de Pertenza

Buscador de vogais en texto:
- Pide unha frase ao usuario.
- Usa `in` para verificar se contén cada vogal.
- Usa `not in` para verificar se non contén números.
- Conta cantas vogais diferentes ten.

```python
frase = input("Introduce unha frase: ").lower()
print("Frase introducida:", frase)
vogais = "aeiou"
cont_vogais = 0

for v in vogais:
    if v in frase:
        cont_vogais += 1
        print(f"A frase contén a vogal '{v}'")

if all(ch not in "0123456789" for ch in frase):
    print("Non contén números")
else:
    print("Contén números")

print("Número de vogais diferentes:", cont_vogais)
```

### Exercicio 8: Walrus

Validador de contrasinal con walrus:
- Pide contrasinal ata que cumpra requisitos.
- Lonxitude >= 8 caracteres.
- Contén maiúscula e número.
- Usa walrus para non repetir len(contrasinal).

```python
while True:
    contrasinal = input("Introduce tu contraseña: ")
    
    if (lonx := len(contrasinal)) < 8:
        print(f"Contraseña demasiado corta ({lonx} caracteres). Necesitas al menos 8.")
        continue

    if not any(c.isupper() for c in contrasinal):
        print("La contraseña debe contener al menos una letra mayúscula.")
        continue

    if not any(c.isdigit() for c in contrasinal):
        print("La contraseña debe contener al menos un número.")
        continue

    print("Contraseña válida.")
    break
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
peso = float(input("Introduce o peso en kg: "))
altura = float(input("Introduce a altura en m: "))

imc = peso / altura**2
print(f"IMC: {imc:.2f}")

if imc < 18.5:
    categoria = "Baixo peso"
elif 18.5 <= imc <= 24.9:
    categoria = "Normal"
elif 25 <= imc <= 29.9:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidade"

print("Categoría:", categoria)
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

```python
texto = input("Introduce un texto: ").lower()
if texto:
    print("Texto introducido:", texto)
    print("Lonxitude:", len(texto))
    print("Contén 'python'?:", "python" in texto)
    vogais = sum(1 for c in "aeiou" if c in texto)
    print("Número de vogais diferentes:", vogais)
    palindromo = texto.replace(" ", "") == texto.replace(" ", "")[::-1]
    print("É palíndromo?:", palindromo)
else:
    print("Texto é None")
```

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
import random

while True:
    secreto = random.randint(1, 100)
    intentos = 0
    while (adivina := int(input("Adiviña o número (1-100): "))):
        intentos += 1
        if adivina < secreto:
            print(f"Número introducido {adivina}. É maior")
        elif adivina > secreto:
            print(f"Número introducido {adivina}. É menor")
        else:
            print(f"O número era {secreto}. Acertaches en {intentos} intentos!")
            break
    if input("Queres xogar outra vez? (si/no): ").lower() != "si":
        break
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
email = input("Introduce o teu email: ")
dominios_validos = ["gmail.com", "hotmail.com", "outlook.com"]

if "@" in email and "." in email.split("@")[-1] and len(email) >= 5 and " " not in email:
    dominio = email.split("@")[-1]
    if dominio in dominios_validos:
        print("Email válido")
    else:
        print("Dominio non válido")
else:
    print("Email non válido")
```
